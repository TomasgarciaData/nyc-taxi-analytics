```python
from pyspark.sql import functions as F
from pyspark.sql.types import *
```

SILVER


```python
#snake_case and change date data type
df_weather = spark.read.table("weather_bronze")

df_weather = df_weather.withColumnRenamed("time", "date") \
    .withColumnRenamed("temperature_2m_max", "temp_max") \
    .withColumnRenamed("temperature_2m_min", "temp_min") \
    .withColumnRenamed("precipitation_sum", "precipitation") \
    .withColumnRenamed("snowfall_sum", "snowfall") \
    .withColumnRenamed("wind_speed_10m_max", "wind_speed_max")

df_weather = df_weather.withColumn("date", F.to_date("date", "yyyy-MM-dd"))

df_weather.write.format("delta").mode("overwrite").saveAsTable("weather_silver")
```
