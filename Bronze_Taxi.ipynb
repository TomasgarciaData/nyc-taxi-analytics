```python
from pyspark.sql import functions as F
from pyspark.sql.types import *

```

Bronze

```python
#read trips
df_trips = spark.read.parquet("Files/bronze/trips")
df_trips.printSchema()
df_trips.count()
```

```python
#read zones
df_zones = spark.read.option("header", True).csv("Files/bronze/zones")
df_zones.printSchema()
df_zones.count()
```

```python
#save delta tables
df_trips.write.format("delta").mode("overwrite").saveAsTable("trips_bronze")
df_zones.write.format("delta").mode("overwrite").saveAsTable("zones_bronze")
```
